```python
y = np.array([1.0,1.5,2.0,2.5,3.0])
data = np.array([8302,19309,22621,12693,7685])
```

```python
plt.errorbar(y-0.5, data, xerr=0.25, fmt='.', color='b', ecolor='y', capsize=0)
plt.plot(y-0.5,data,c='#f99157')
plt.xlabel('y')
plt.ylabel('No. of particles')
```

## Model: BDT (Gradient Boost)

### Configuration 1
<b>Details:</b>
* No. of trees - 1000
* min. node size - 2.5%
* shrinkage - 0.1
* Bagged sample fraction - 0.5
* nCuts - 20
* max depth - 5
* Seperation type - Misclassification Error

```python
eff1 =  np.array([0.9169,0.936,0.9649,0.9781,0.9883])
pur1 = np.array([0.97,0.985,0.994,0.997,0.998])
model_acc1 = np.array([0.976,0.989,0.999,1.0,1.0])
err = 1-model_acc1

# Plotting
plt.plot(y,eff1,c='g',label="efficiency")
plt.errorbar(y, eff1, yerr=err/2, fmt='o', color='b', ecolor='y', capsize=5)
plt.plot(y,pur1,c='c',label="purity")
plt.errorbar(y, pur1, yerr=err/2, fmt='o', color='k', ecolor='r', capsize=5)
plt.xlabel('y')
plt.ylabel('Signal')
#plt.ylim([0,1.03])
plt.title("Model: BDTG; config-1")
plt.legend()
```
