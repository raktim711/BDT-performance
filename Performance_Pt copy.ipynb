## Model: BDT (Gradient Boost)

### Configuration 1
<b>Details:</b>
* No. of trees - 1000
* min. node size - 2.5%
* shrinkage - 0.1
* Bagged sample fraction - 0.5
* nCuts - 20
* max depth - 5
* Seperation type - Misclassification Error

```python
eff1 =  np.array([0.9388,0.9573,0.97640,0.9854])
pur1 = np.array([0.989,0.992,0.996,0.998])
model_acc1 = np.array([0.989,0.996,1.0,1.0])
err = 1-model_acc1

# Plotting
plt.plot(pt,eff1,c='g',label="efficiency")
plt.errorbar(pt, eff1, yerr=err/2, fmt='o', color='b', ecolor='y', capsize=5)
plt.plot(pt,pur1,c='c',label="purity")
plt.errorbar(pt, pur1, yerr=err/2, fmt='o', color='k', ecolor='r', capsize=5)
plt.xlabel(r'$P_T$')
plt.ylabel('Signal')
#plt.ylim([0,1.03])
plt.title("Model: BDTG; config-1")
plt.legend()
```
